```python
import matplotlib.pyplot as plt
import numpy as np

N=100
x=np.empty(N)
x1=np.empty(N)
x2=np.empty(N)
xg=np.empty(N)
dxupg=np.empty(N)
dxumg=np.empty(N)

b=2.5;
c=50
sx=1.0;
xmin=0.0;
m=0.8;
K1=np.tanh(b*(m-1))
K2=np.tanh(b*m)
for i in range(100):
    sxu1= (sx*i)/(N-1)
    x1[i] = xmin + ((np.tanh(b*(sxu1+m)))-K1)/(K2-K1)
    sxu2= sx*(i+1)/(N-1)
    x2[i] = xmin + ((np.tanh(b*(sxu2+m)))-K1)/(K2-K1)
    xg[i]  =x1[i]+ 0.5*(x2[i]-x1[i])
    dxupg[i] = x2[i]-xg[i]
    dxumg[i] = xg[i]-x1[i]
    x[i]=i

plt.figure(0)    
plt.plot(x,x1)

plt.figure(1)    
plt.plot(x,x2)

plt.figure(2)    
plt.plot(x,xg) 

plt.figure(3)    
plt.plot(x,dxupg)

plt.figure(4)    
plt.plot(x,dxumg) 
```
