```python
import matplotlib.pyplot as plt
import numpy as np

N=100
x=np.empty(N)
x1=np.empty(N)
x2=np.empty(N)
xg=np.empty(N)
dxupg=np.empty(N)
dxumg=np.empty(N)

deltax=25.2;
sx=1.0;
xmin=0.0;
for i in range(100):
    sxu1= (sx*i)/(N-1)
    x1[i] = xmin + (1.0 + (np.tanh(deltax*(sxu1-1)/2.0))/(np.tanh(deltax/2.0))) 
    sxu2= sx*(i+1)/(N-1)
    x2[i] = xmin + (1.0 + (np.tanh(deltax*(sxu2-1)/2.0))/(np.tanh(deltax/2.0))) 
    xg[i]  =x1[i]+ 0.5*(x2[i]-x1[i])
    dxupg[i] = x2[i]-xg[i]
    dxumg[i] = xg[i]-x1[i]
    x[i]=i

plt.figure(0)    
plt.plot(x,x1)

plt.figure(1)    
plt.plot(x,x2)

plt.figure(2)    
plt.plot(x,xg) 

plt.figure(3)    
plt.plot(x,dxupg)

plt.figure(4)    
plt.plot(x,dxumg) 
```
